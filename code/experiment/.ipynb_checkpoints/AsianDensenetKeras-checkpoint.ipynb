```python
def reduced_densenet(input_shape: tuple=(128, 128, 3), classes: int=10, optimizer: (str, Optimizer)='adam'):
    inputs = Input(shape=input_shape)
    x = dense_block(inputs, 32, 2)
    x = MaxPooling2D()(x)
    x = transition_block(x, 32)
#     x = dense_block(x, 64, 2)
    x = MaxPooling2D()(x)
    x = transition_block(x, 64)
    x = Flatten()(x)
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.5)(x)
    x = Dense(10, activation='softmax')(x)
    model = Model(inputs, x)
    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy'])
    return model
```

```python
model = reduced_densenet(optimizer=Adam(lr=1e-4))
model.summary()
model, history = train_model(model, 200)
```

```python
plot_history(history)
```

```python
save_model(model, history, 'densenet_keras_reduced')
```
