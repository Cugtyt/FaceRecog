```python
def smooth(line: list, window: int):
    """Smooth line with respect to given window.
    
    Args:
        line: list, data inputs
        window: int, smooth window
    Returns:
        result: list, smoothed line
    """
    half = window // 2
    result = line[:half]
    for i in range(half, len(line) - half):
        result.append(np.mean(line[i - half: i + half]))
        
    return result
```

```python
def plot_acc_loss(history: dict, smooth_factor: int, alpha: float, title: dict, label: dict, legend: dict):
    sns.set()
    sns.set_style('whitegrid')
    sns.despine()
    
    fig, axarr = plt.subplots(1, 2, figsize=(16, 5))
    
    acc_smooth = smooth(history['acc'], smooth_factor)
    val_acc_smooth = smooth(history['val_acc'], smooth_factor)
    loss_smooth = smooth(history['loss'], smooth_factor)
    val_loss_smooth = smooth(history['val_loss'], smooth_factor)
    
    axarr[0].plot(history['acc'], alpha=alpha)
    axarr[0].plot(acc_smooth, label=legend['acc'])
    axarr[0].plot(history['val_acc'], alpha=alpha)
    axarr[0].plot(val_acc_smooth, label=legend['val_acc'])
    axarr[0].set_title(title['acc'])
    axarr[0].set_xlabel(label['acc_x'])
    axarr[0].set_ylabel(label['acc_y'])
    axarr[0].legend()

    axarr[1].plot(history['loss'], alpha=alpha)
    axarr[1].plot(loss_smooth, label=legend['loss'])
    axarr[1].plot(history['val_loss'], alpha=alpha)
    axarr[1].plot(val_loss_smooth, label=legend['val_loss'])
    axarr[1].set_title(title['loss'])
    axarr[1].set_xlabel(label['loss_x'])
    axarr[1].set_ylabel(label['loss_y'])
    axarr[1].legend()
```
